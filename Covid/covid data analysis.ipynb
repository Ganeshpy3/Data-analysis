```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("covid_19_data.csv")
```

```python
data.head()
```

```python
data.sort_values("Confirmed",ascending=False)
```

```python
india=data[data["Country/Region"]=="India"]
```

```python
india.tail()
```

```python
india.sort_values("Confirmed",ascending=False)
```

```python
tn=india[india["Province/State"]=="Tamil Nadu"]
```

```python
tn.describe()
```

```python
tn.info()
```

```python
mh=india[india["Province/State"]=="Maharashtra"]
```

```python
tn.plot(kind="Line",x="ObservationDate",y="Confirmed",figsize=(20,10))
mh.plot(kind="Line",x="ObservationDate",y="Confirmed")
```

```python
plt.figure(figsize=(20,10))
plt.plot(tn["ObservationDate"],tn["Confirmed"],label="Tamil Nadu")
plt.plot(mh["ObservationDate"],mh["Confirmed"],label="Maharastra")
plt.legend()
```
