```python
from bs4 import BeautifulSoup
import  json
import requests
import time
import pandas as pd
```

```python
crypto={"loopring":["priceValue","sc-15yy2pl-0 feeyND"],"ethereum":["priceValue","sc-15yy2pl-0 gEePkg"],
       "bitcoin":["priceValue","sc-15yy2pl-0 feeyND"],"dogecoin":["priceValue","sc-15yy2pl-0 feeyND"]}
```

```python
def show_coins(crypto):
    df=None
    
    for i in crypto:
        URL="https://coinmarketcap.com/currencies/"+i+"/"
        HEADERS = ({'User-Agent':
                'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
            'Accept-Language': 'en-US'})
        webpage = requests.get(URL,headers=HEADERS)
        state_soup = BeautifulSoup(webpage.content, "lxml")
        price=state_soup.findAll("div",{"class":crypto[i][0]})
        percent=state_soup.findAll("span",{"class":crypto[i][1]})
        name=i.upper()
        for i in price:
            pric=i.text.split(" ")[0]
#             print("Price  :{}".format(i.text.split(" ")[0]))
            break
        for i in percent:
            per=i.text.split(" ")[0]
#             print("percent:{}".format(i.text.split(" ")[0]))
            break
        text=str(percent)
        if(text.find("icon-Caret-down")<0):
            colour="Green"
        else:
            colour="RED"
        if df is not None:
            df=pd.concat([df,pd.DataFrame([{"Name":name,"price":str(pric),"up/down":colour,"percentage":per}])])
        else:
            df=pd.DataFrame([{"Name":name,"price":str(pric),"up/down":colour,"percentage":per}])
    return df
```

```python
df=None
i=0
while True:
    if df is None:
        df=show_coins(crypto)
    else:
        df=pd.concat([df,show_coins(crypto)])
    i=i+1
    if(i>10):
        break
```
