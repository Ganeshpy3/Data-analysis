```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# removing age less than 7 and greater than 90

```python
data=data[(data["Age"]>=7)&(data["Age"]<=90)]
```

```python
data.reset_index(drop=True)
```

```python
data.describe()
```

```python
data.columns
```

# India coders favourite programing language

```python
india=data[data["Country"]=="India"]
```

```python
india.dropna(inplace=True)
```

```python
prolang={}
for i in india["LanguageWorkedWith"]:
    for j in i.split(';'):
        if j in prolang:
            prolang[j]+=1
        else:
            prolang[j]=1
```

```python
prolang
```

```python
indiacoderslang=pd.DataFrame(prolang,index=["programing_languages"]).T
```

```python
indiacoderslang.sort_values(by="programing_languages",ascending=True,inplace=True)
indiacoderslang.tail(5)
```

```python
plt.figure(figsize=(20,10))
plt.barh(indiacoderslang.index,indiacoderslang["programing_languages"])
plt.title("Stackoverflow 2020 Indian coders most used language")
```

```python
countlang={}
for i in data["LanguageWorkedWith"].dropna():
    for j in i.split(';'):
        if j in countlang:
            countlang[j]+=1
        else:
            countlang[j]=1
```

```python
countlang
```

```python
worldcoderslang=pd.DataFrame(countlang,index=["programing_languages"]).T
```

```python
worldcoderslang.sort_values(by="programing_languages",ascending=True,inplace=True)
worldcoderslang.tail(5)
```

```python
plt.figure(figsize=(20,10))
plt.barh(worldcoderslang.index,worldcoderslang["programing_languages"])
plt.title("Stackoverflow 2020 world coders most used language")
```
