# india coders


```python
india=data[data["Country"]=="India"]
```

```python
india.dropna(inplace=True)
```

```python
prolang={}
for i in india["LanguageWorkedWith"]:
    for j in i.split(';'):
        if j in prolang:
            prolang[j]+=1
        else:
            prolang[j]=1
```
