```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("match.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data["venue"].unique()
```

```python
chennai=data[data["venue"]=="MA Chidambaram Stadium, Chepauk"]
```

```python
chennai
```

```python
chennai_1stinnings=chennai[chennai["innings"]==1]
```

```python
chennai_2ndinnings=chennai[chennai["innings"]==2]
```

```python
f_innings_score=chennai_1stinnings.groupby("match_id").sum()["runs_off_bat"]
s_innings_score=chennai_2ndinnings.groupby("match_id").sum()["runs_off_bat"]
```

```python
f_innings_score.plot(kind="Bar",figsize=(20,10))
```

```python
s_innings_score.plot(kind="Bar",figsize=(20,10))
```

```python
s_innings_score
```

```python
f_innings_score
```

```python
s_innings_score=s_innings_score.reset_index()
s_innings_score=s_innings_score.drop(["match_id"],axis=1)
```

```python
f_innings_score=f_innings_score.reset_index()
f_innings_score=f_innings_score.drop(["match_id"],axis=1)
```

```python
second_innings = s_innings_score.set_axis(["second_innings"], axis=1)
f_innings_score.set_axis(["first_innings"], axis=1)
```

```python
s=np.array(s_innings_score)
b=[]
for i in s:
    for j in i:
        b.append(j)
```

```python
s=np.array(f_innings_score)
a=[]
for i in s:
    for j in i:
        a.append(j)
```

```python
figure_size = (20, 10)
plt.figure(figsize=figure_size)
plt.plot(f_innings_score,'r-',label='First Innings')
plt.plot(s_innings_score,'b',label="Second Innings")
plt.title('Chennai Ground')
plt.legend()
plt.show()
```

```python
first_win=0
second_win=0
for i in range(len(s_innings_score)):
    if(s_innings_score[i:i+1].values>f_innings_score[i:i+1].values):
        second_win +=1
    else:
        first_win +=1
    
```
